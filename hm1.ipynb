# Задание 2.

```python
from string import punctuation
from nltk.tokenize import word_tokenize
from nltk.stem.snowball import SnowballStemmer
```

```python
with open('gr') as f:
    text = f.read()

punct = '[{}«»—…“”–]'.format(punctuation)

text_nopun = re.sub(punct, '' ,text.lower())

tokens = word_tokenize(text_nopun)
tokens[:10]
```

```python
s = SnowballStemmer('russian')
stems = [(word, s.stem(word)) for word in tokens]
stems[:5]
```

### 1. Два разных слова ошибочно свелись к одинаковой основе

```python
temp = []
eq = set()
for stem in stems:
    if stem[1] in temp:
        eq.add(stem)
    else:
        temp.append(stem[1])
sorted(eq, key=lambda x: x[1])
```

Следующие разные слова свелись:  
вид, видеть -> вид  
какой, как -> как  
целый, цель -> цел  
Сложно назвать такие случаи ошибкой, так как основы действительно совпадают. Так что можно сказать, что ошибок такого типа нет.

### 2. Две словоформы одного слова после стемминга отличаются друг от друга

```python
eq
```

выраженные, выражено ->  выражен, выраж

### 3. Слово не изменилось после стемминга

```python
set([
    stem for stem in stems \
    if stem[0] == stem[1] and len(stem[0]) > 4
])
```

```python
lemmas = [
    [
        pmrph.parse(token)[0].normal_form for token in sent
    ] for sent in sent_tokens
]
print(lemmas)
```
