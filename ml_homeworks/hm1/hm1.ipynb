```python
import pandas as pd
```

```python
weather_hist = pd.read_csv('weatherHistory.csv')
```

```python
weather_hist['date'] = pd.to_datetime(weather_hist['Formatted Date'], utc=True)
weather_hist.drop(columns=['Formatted Date', 'Loud Cover'], inplace=True)
weather_hist.head()
```

```python
weather_hist['month'] = [date.month for date in weather_hist['date']]
weather_hist['year'] = [date.year for date in weather_hist['date']]
month_table = weather_hist.groupby('month').mean()
month_table[month_table['Wind Speed (km/h)'] == month_table['Wind Speed (km/h)'].max()]
```

```python
year_table = weather_hist.groupby('year').mean()
year_table[year_table['Humidity'] == year_table['Humidity'].max()]
```

```python
weather_hist['week'] = weather_hist['date'].dt.week
week_table = weather_hist.groupby('week').var()
week_table[week_table['Temperature (C)'] == week_table['Temperature (C)'].max()]
```
